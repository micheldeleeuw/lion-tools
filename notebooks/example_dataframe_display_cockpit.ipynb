```python
import lion_tools
import ipynbname
spark=lion_tools.start_spark()
lion_tools.extend_dataframe()
import os
import shutil
```

```python
my_df = spark.createDataFrame([[1, "Alice]\\n in hell{}"], [2, "Bob"]], ["id", "name"])
spark.sql("DROP TABLE IF EXISTS my_temp_table")

table_path = os.path.join(spark.conf.get("spark.sql.warehouse.dir"), 'my_temp_table').replace("file:/", "/")
shutil.rmtree(table_path, ignore_errors=True)
my_df.write.mode("overwrite").saveAsTable("my_temp_table")

# my_df.eC(lazy=False, sort=-1)
# my_df.eC('my cool\'" dataframe', lazy=False, sort=-1)
spark.table("my_temp_table").eC()

from lion_tools import Cockpit
# Cockpit.print_status()
Cockpit.run()
```
