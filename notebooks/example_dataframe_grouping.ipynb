```python
import ipynbname
import pyspark.sql.functions as F
import lion_tools
spark=lion_tools.get_or_create_spark()
p=6
df = spark.read.parquet(str(ipynbname.path().parent.parent.joinpath("datasets", "movies.parquet")))
df = (
    df
    .withColumn('piv', (F.floor(F.rand() * 3) + 1).cast('int'))
    .withColumn('piv', F.expr("if(piv = 1, null, piv)"))
)
```

```python
# single measure, default
(
    df
    # .displayp()
    .withColumn('Distributor', F.expr('substring(Distributor, 1, 1)')) # type: ignore
    .filter("Distributor in ('A', 'B', 'C', 'D')") # type: ignore
    # .displayp()
    .group('Distributor') # type: ignore
    .pivot('MPAA Rating', 'piv', pivot_type='default') # type: ignore
    .totals()
    .agg('count(distinct ifnull(Director, "null"))') # type: ignore
    # .display(p=p, compact=1)
    .show()
)
```

```python
# single measure, default
(
    df
    # .displayp()
    .withColumn('Distributor', F.expr('substring(Distributor, 1, 1)')) # type: ignore
    .filter("Distributor in ('A', 'B', 'C', 'D')") # type: ignore
    # .displayp()
    .group('Distributor') # type: ignore
    .pivot('MPAA Rating', 'piv', pivot_type='default') # type: ignore
    .totals()
    .agg('count(distinct ifnull(Director, "null"))') # type: ignore
    .display(p=p, compact=1)
    # .show()
)
```

```python
# single measures, column_grouped
(
    df
    .withColumn('Distributor', F.expr('substring(Distributor, 1, 1)')) # type: ignore
    .filter("Distributor in ('A', 'B', 'C', 'D')") # type: ignore
    .group('Distributor') # type: ignore
    .pivot('MPAA Rating', 'piv', pivot_type='column_grouped')
    .pivot_totals(grand_total=True)
    .totals()
    .agg('count(*) as `count`') # type: ignore
    .display(p=p, compact=1)
)
```

```python
# multiple measures, column_grouped
(
    df
    .withColumn('Distributor', F.expr('substring(Distributor, 1, 1)')) # type: ignore
    .filter("Distributor in ('A', 'B', 'C', 'D')") # type: ignore
    .group('Distributor') # type: ignore
    .pivot('MPAA Rating', 'piv', pivot_type='column_grouped')
    .totals()
    .agg('count(distinct Director)', 'count(*) as `count`', 'avg(`IMDB Rating`)', 'sum(`IMDB Votes`)') # type: ignore
    .display(p=p, compact=1)
)
```

```python
# single measure, pivot_grouped
(
    df
    .withColumn('Distributor', F.expr('substring(Distributor, 1, 1)')) # type: ignore
    .filter("Distributor in ('A', 'B', 'C', 'D')") # type: ignore
    .group('Distributor') # type: ignore
    .pivot('MPAA Rating', 'piv', pivot_type='pivot_grouped')
    .totals()
    .agg('count(distinct Director)') # type: ignore
    .display(p=p, compact=1)
)
```

```python
# multiple measures, pivot_grouped
(
    df
    .withColumn('Distributor', F.expr('substring(Distributor, 1, 1)')) # type: ignore
    .filter("Distributor in ('A', 'B', 'C', 'D')") # type: ignore
    .group('Distributor') # type: ignore
    .pivot('MPAA Rating', 'piv', pivot_type='pivot_grouped')
    .totals()
    .agg('count(distinct Director)', 'avg(`IMDB Rating`)', 'sum(`IMDB Votes`)') # type: ignore
    .display(p=p, compact=1)
)
```
