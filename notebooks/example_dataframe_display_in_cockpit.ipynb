```python
import lion_tools
import ipynbname
import time
import os, shutil
import datetime

spark=lion_tools.get_or_create_spark()
lion_tools.extend_dataframe()
lion_tools.Cockpit.clear()
lion_tools.CockpitLogging.redirect()
```

```python

my_df = spark.createDataFrame([[1, "Alice]\\n in hell{}"], [2, "Bob"]], ["id", "name"])
my_df.eC(f'{datetime.datetime.now()}', lazy=False, page_length=10)
print('Alice bob sent to cockpit')
# lion_tools.start_cockpit(0.05)
```

```python
df = spark.read.parquet(str(ipynbname.path().parent.parent.joinpath("datasets", "movies.parquet")))
spark.sql('drop table if exists default.movies')
table_path = os.path.join(spark.conf.get("spark.sql.warehouse.dir"), 'movies').replace("file:/", "/")
shutil.rmtree(table_path, ignore_errors=True)
df.write.mode("overwrite").saveAsTable("default.movies")
spark.table('default.movies').join(spark.table('default.movies')).eSources()
```
