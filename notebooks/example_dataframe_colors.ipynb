```python
import ipynbname
movies = spark.read.parquet(str(ipynbname.path().parent.parent.joinpath("datasets", "movies.parquet")))

movies = movies.select(*[col for col in movies.columns[0:5]])

movies.eSetColors(
    {'column': 'Director', 'color': 'red', 'style': 'bold'},
    {'columns': ['Director', 'Distributor'], 'color': '#123456'},
    {'color': 'red', 'condition': '`IMDB Rating` < 6', 'styles': ['bold']},
    {'columns': ['IMDB Votes'], 'color': '1', 'styles': ['italic']},
).eD()

# movies.eSetColors(
#     {'color': 'red', 'condition': F.col('IMDB Votes') < 6},
# ).show(truncate=False)
```
