```python
import ipynbname
movies = spark.read.parquet(str(ipynbname.path().parent.parent.joinpath("datasets", "movies.parquet")))

movies = movies.select(*[col for col in movies.columns[0:5]])

movies.eSetColors(
    {'column': 'Director', 'color_code': 'traffic_light_red', 'style_code': 'bold'},
    {'columns': ['Director', 'Distributor'], 'color_code': 'traffic_light_green', 'style_code': 'italic'},
    {'color_code': 'traffic_light_red', 'condition': '`IMDB Rating` < 6', 'style_code': 'bold'},
    {'columns': ['IMDB Votes'], 'color_code': 'traffic_light_yellow', 'style_code': 'italic'},
).eRound(1).eD()

# import pyspark.sql.functions as F
# movies.eSetColors(
#     {'color_code': 'traffic_light_red', 'condition': F.col('IMDB Votes') < 6},
# ).show(truncate=False)
```
